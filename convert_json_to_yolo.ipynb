```python
import json, os
dataset_path = f"/Users/wenjie.zhang/Library/CloudStorage/OneDrive-Personal/OneDrive - National University of Singapore/Lab/dataset/dataset-bought/dataset/28绝缘子闪络、破损等缺陷检测数据集/Train"
json_filename= f"labels_v1.2.json"
json_path = os.path.join(dataset_path, json_filename)
# Load the JSON data from the file
with open(json_path) as f:
    data = json.load(f)
```

```python
# clean the path
import os, glob
map(os.remove, glob.glob(os.path.join(dataset_path, "*")))
```

```python
# remember to clean the path before running
os.system(f"rm -r {dataset_path}/yolo_images/")
os.system(f"rm -r {dataset_path}/labels/")
os.system(f"mkdir {dataset_path}/yolo_images/")
os.system(f"mkdir {dataset_path}/yolo_images/train/")
os.system(f"mkdir {dataset_path}/yolo_images/val/")
os.system(f"mkdir {dataset_path}/labels/")
os.system(f"mkdir {dataset_path}/labels/train/")
os.system(f"mkdir {dataset_path}/labels/val/")
import cv2, numpy as np
labels = []
val_portion = 0.2
for file_info in data:
    filename = file_info['filename']
    is_val = np.random.uniform(0, 1) < val_portion
    # read the image named as filename
    img = cv2.imread(os.path.join(dataset_path, f"Images/{filename}"))
    img_height, img_width, _ = img.shape
    filename = filename.replace('.JPEG', '.jpg').replace('.JPG','.jpg').replace('.jpeg','.jpg')
    if is_val:
        # save the image to val
        cv2.imwrite(os.path.join(dataset_path, f"yolo_images/val/{filename}"), img)
    else:
        # save the image to train
        cv2.imwrite(os.path.join(dataset_path, f"yolo_images/train/{filename}"), img)
    for obj in file_info['Labels']['objects']:
        # Extract the class label, bounding box coordinates, and image size
        label = obj['name']+f"-string:{obj['string']}"+f"-conditions:{obj.get('conditions', 'None')}"
        x = obj['bbox'][0]
        y = obj['bbox'][1]
        width = obj['bbox'][2]
        height = obj['bbox'][3]
        # Calculate the center coordinates and the normalized width and height
        # relative to the image size
        x_center = (x + width / 2) / img_width
        y_center = (y + height/ 2) / img_height
        width_norm = width / img_width
        height_norm = height / img_height
        # Generate the YOLO format output
        if label not in labels:
            labels.append(label)
        # get the index of the label in the labels
        label_index = labels.index(label)
        output = ' '.join([str(label_index), str(x_center), str(y_center),
                        str(width_norm), str(height_norm)])
        # seperate train and val according to the val_portion, if random uniform is less than val_portion, save to val
        if is_val:
            with open(os.path.join(dataset_path, f"labels/val/{filename.split('.')[0]}.txt"), 'a') as f:
                f.write(output + '\n')
        else:
            with open(os.path.join(dataset_path, f"labels/train/{filename.split('.')[0]}.txt"), 'a') as f:
                f.write(output + '\n')
print(labels)
# save labels to yolo-classes.txt
with open(os.path.join(dataset_path, f"yolo-classes.txt"), 'w') as f:
    f.write(str(labels))
len(labels)
```
